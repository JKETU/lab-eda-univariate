```python
mean_price = df['price'].mean()
mean_price
```

```python
median_price = df['price'].median()
median_price
```

```python
mode_price = df['price'].mode()
mode_price
```

```python
# Calculate variance, standard deviation, range, and interquartile range (IQR) for the 'price' column
variance_price = df['price'].var()
std_dev_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

# Display the results
print(f"Variance of price: {variance_price}")
print(f"Standard deviation of price: {std_dev_price}")
print(f"Range of price: {range_price}")
print(f"Interquartile range (IQR) of price: {iqr_price}")
```

```python
# Plotting a histogram to visualize the distribution of product prices
plt.figure(figsize=(8, 8))
plt.hist(df['price'], bins=50, edgecolor='black', color='skyblue')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plotting a box plot to showcase the spread and potential outliers in product pricing
plt.figure(figsize=(5, 5))
plt.boxplot(df['price'], vert=False, patch_artist=True)
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.grid(True)
plt.show()
```

```python
# Part 3: Unpacking Product Ratings
#Business Question: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?
```

```python
#Measures of Centrality:

#Calculate the mean, median, and mode for the rating of products.
#How do customers generally rate products? Is there a common trend?

df['stars']
```

```python
mean_rating = df['stars'].mean ()
median_rating = df['stars'].median ()
mode_rating = df['stars'].mode ()

print (f'mean product rating : {mean_rating}')
print (f'median product rating : {median_rating}')
print (f'mode product rating : {mode_rating}')
```

```python
#2. Measures of Dispersion

variance_rating = df['stars'].var ()
std_dev_rating = df['stars'].std ()
quartile_rating = df['stars'].quantile (0.75) - df['stars'].quantile (0.25) #interquartile range

print (f'variance of product rating : {variance_rating}')
print (f'standard deviation of product rating : {std_dev_rating}')
print (f'interquartile of product rating : {quartile_rating}')
```

```python
#3. Shape of the Distribution

# Calculate the skewness and kurtosis for the 'rating' column
skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurtosis()

# Display the results
print(f"Skewness of rating: {skewness_rating}")
print(f"Kurtosis of rating: {kurtosis_rating}")
```

```python
# Plotting a histogram to visualize the distribution of product ratings
plt.figure(figsize=(10, 6))
plt.hist(df['stars'], bins=20, edgecolor='black', color='lightgreen')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
